```python
import pandas as pd
```

```python
path = "./exercise_data/chipotle.tsv"
```

```python
chipo = pd.read_csv(path, sep='\t')
```

```python
# 查看前10行内容
chipo.head(10)
```

```python
# 数据集中有多少个列(columns)
chipo.shape[1]
```

```python
chipo.item_price = chipo.item_price.apply(lambda x: float(x[1:]))
```

```python
chipo.head(3)
```

```python
# 查询商品种类的数量
print('Total items: ', len(chipo.item_name.unique()))
print('Total items: ', chipo.item_name.value_counts().count())
print('Total items: ', len(chipo.groupby(['item_name']).groups))
```

```python
# 获取销量最多的商品名
chipo.item_name.value_counts().head().index
```

```python
# 获取销量最多的商品销售量
chipo.item_name.value_counts().head()
```

```python
# 查询最受欢迎的食物及其销量 , 销售额
## 查询最受欢迎的食物
favorite_items = chipo.groupby(['item_name']).sum().sort_values('quantity',ascending=False)
## 删除order_id 列
favorite_items = favorite_items.drop('order_id', axis=1)
favorite_items.head()
```

```python
# 查询最受欢迎食物（下单数， 销量， 销售额）
favorite_items = chipo.groupby(['item_name']).agg({'order_id': 'count', 'quantity':'sum','item_price':'sum'}).sort_values('quantity',ascending=False)
favorite_items.head()
```

```python
# 查询订单的商品总数量
chipo.quantity.sum()
```

```python
# 统计总收入
chipo.item_price.sum()
```

```python
# 统计订单总数
chipo.order_id.value_counts().count()
```
